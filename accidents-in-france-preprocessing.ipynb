### Handling Missing Values

```python
def missing_vals(df):
    for column in df.columns :
        null_count = df.isnull()[column].sum()
        
        if null_count == 0:
            continue
        
        total_count = df[column].shape[0]
        percentage = round((null_count/total_count)*100, 2)
        
        print(f'percentage of missing vals in {column} : {percentage}')
```

### Feature Importance

##### Feature Importance refers to techniques that calculate a score for all the input features for a given model — the scores simply represent the “importance” of each feature. A higher score means that the specific feature will have a larger effect on the model that is being used to predict a certain variable.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, MinMaxScaler
from sklearn.metrics import classification_report
```

```python
def transformation(column):
  max_value = column.max()
  sin_values = [np.sin((2*22*int(x))/(int(max_value)*7)) for x in list(column)]
  cos_values = [np.cos((2*22*int(x))/(int(max_value)*7)) for x in list(column)]
  return sin_values, cos_values
```

```python
hours_sin, hours_cos = transformation(df.hours)
```
